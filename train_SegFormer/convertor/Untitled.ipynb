```python
import numpy as np
img2 = cv2.imread(img_path)
uniques, counts = np.unique(img2[::,], return_counts=True)
color_seg = np.zeros((img2.shape[0], img2.shape[1]), dtype=np.uint8) # height, width, 3

new_map = [[0 for _ in range(img2.shape[1])] for __ in range(img2.shape[0])]
unique = []
for i, line in enumerate(img2):

    for j, dot in enumerate(line):     

        if dot[0]== 127:
            color_seg[i][j] = 255
        else : 
            pass
            color_seg[i][j] = 0

            

from imantics import Polygons, Mask
plain_polygons = Mask(color_seg).polygons()
```

```python
draw_img = np.zeros((img2.shape[0], img2.shape[1],3), dtype=np.uint8) # height, width, 3
draw_img[draw_img==0] = 255
cnt = 0 
for polygon in plain_polygons.segmentation:
#     print(polygon)
    if len(polygon) <3 :
        continue
    cnt += 1
    points = np.array(polygon)
    points = points.reshape(int(len(polygon)/2),2)

    color_np = np.asarray(np.random.choice(range(256), size=3), dtype=np.uint8)
    color_fill = (color_np[0].item(),color_np[1].item(),color_np[2].item())
    image = cv2.fillPoly(draw_img, pts=[points], color=color_fill)
plt.figure(figsize=(15, 10))
plt.imshow(draw_img)
plt.show()
print(cnt)
```
