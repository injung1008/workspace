```python
import os
import json
json_label_path = './train/json/'
total_j_dict = {}
fileEx = r'.json'
json_list = [file for file in os.listdir(json_label_path) if file.endswith(fileEx)]
for j_file in json_list :
    json_file_path = f'{json_label_path}/{j_file}'
    with open (json_file_path, "r") as f:
        j_dict = json.load(f)
        total_j_dict[j_file] = j_dict
print(total_j_dict)
```

```python
total_cate_id_dict = {}
print(f'total file : {len(total_j_dict)}')
for json_file in total_j_dict :
    for anno in total_j_dict[json_file]['annotation']:
        class_id = anno['class_id']
        class_type = anno['class_type']
        if class_id not in total_cate_id_dict :
            total_cate_id_dict[class_id] = class_type

new_label_dict = {}
for idx, cate in enumerate(total_cate_id_dict) :
    new_label_dict[idx] = total_cate_id_dict[cate]
save_j_path = './train.json'
with open (save_j_path, "w") as f:
    json.dump(new_label_dict,f)
    
print('',total_cate_id_dict)
print('',new_label_dict)

'''
{"0": "obstacle", "1": "vegetation", "2": "void", "3": "truck", "4": "building", "5": "sidewalk", 
"6": "crosswalk", "7": "road", "8": "pedestrian", "9": "stone", "10": "bicycle", "11": "vehicle"}
'''

convert_label_dict = {v:k for k,v in new_label_dict.items()} 
print('', convert_label_dict)
```

```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
image_dir_path = './train/image'
for json_file in total_j_dict :
    
    img_file = json_file.replace('_label.json','.jpg')

    save_img = json_file.replace('_label.json','.png')
    
    img_path = f'{image_dir_path}/{img_file}'
    img = cv2.imread(img_path)
    img[img < 256] = 0

    for anno in total_j_dict[json_file]['annotation']: 
        polygon_list = anno['polygon']
        class_type = anno['class_type']
        label_id = int(convert_label_dict[class_type])
        color_fill = [label_id,label_id,label_id]
#         color_fill = palette[label_id]


        
        polygon = []
        for poly in polygon_list:
            polygon.append(poly[0])
            polygon.append(poly[1])
        
        points = np.array(polygon)
        points = points.reshape(int(len(polygon)/2),2)
        pts = np.array(points,np.int32)
        pts = pts.reshape((-1, 1, 2))

        cv2.fillPoly(img, [pts],color_fill)

    save_dir_path = './train/annotations_rgb'
    save_img_file = f'{save_dir_path}/{save_img}'
    
    print('save',save_img_file)

    convert_pil_image = Image.fromarray(img)
    convert_pil_image = convert_pil_image.convert('L')
    convert_pil_image.save(save_img_file,'png')
#     img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
#     cv2.imwrite(save_img_file, img)
    
# # print('done')
```
