```python
import timm
import torch
inceptionV4 = timm.create_model('inception_v4', pretrained=False, num_classes=3)

#########
# Instantiation
##########

# check model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# x = torch.randn((3, 3, 256,256)).to(device)
model = inceptionV4.to(device)



path2weights = '/DATA_17/ij/trt_inference/make_trt/incep_helmet.pt'

batch_size = 1

input = torch.randn(batch_size,3, 256, 128).cuda()  #배치, 채널, 세로(h), 가로(w)


model.load_state_dict(torch.load(path2weights))
model.eval()

output = model(input)

## pytorch to onnx 
torch.onnx.export(
    model,                                # model being run
    input,    # model input (or a tuple for multiple inputs)
    "incep_helmet.onnx", # where to save the model (can be a file or file-like object)
    verbose=True, 
    opset_version=11,
    input_names = ['input'],              # the model's input names
    output_names = ['output'],
    dynamic_axes = {'input' : {0 : 'batch_size'},
                    'output' : {0 : 'batch_size'}}
)            # the model's output names
```
